# Approssimare un'equazione
In questo notebook utilizzeremo Tensorlfow per realizzare il più semplice dei modelli di Machine Learning, quello che ci permette di approssimare una semplice funzione lineare.

## Importiamo i Moduli

```
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

## Creiamo il Dataset
La funzione che proveremo ad approssimare avrà questa forma: 
$$ y = ax + b $$
nel nostro caso sarà a=2 e b=1, quindi la funzione è:
$$ y = 2x + 1 $$
per farlo abbiamo bisogno di una serie di valori per la x e i corrispondenti valori per la y, inseriamoli all'interno di due array numpy.

```
x = np.array([3, 5, 8, -1, 10, -7, 0, 11])
y = np.array([7, 11, 17, 1, 21, -13, 1, 23])
```

questi due array corrispondono rispettivamente a features e target del nostro modello.

## Creiamo il Modello
Per creare una nuova Rete Neurale possiamo utilizzare la classe Sequential di Tensorflow che accetta in input i vari strati che compongono la Rete. Per questo esercizio dobbiamo addestrare un semplice modello di regressione lineare, quindi sarà sufficente aggiungere un unico strato utilizzando la classe Dense.

```
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

Dopo aver definito il modello dobbiamo configurarlo per la fase di addestramento, possiamo farlo utilizzando il metodo compile.

```
model.compile(optimizer="sgd", loss="mean_squared_error")
```

Infine possiamo avviare l'addestramento con il metodo fit, passando features e target e specificando il numero di epoche.

```
model.fit(x, y, epochs=500)
```

Abbiamo addestrato il nostro modello di Machine Learning, che contiene al suo interno la funzione approssimata, possiamo utilizzarla utilizzando il metodo *predict*.

```
y_pred = model.predict([2])
print(y_pred)
```

Come vedi il metodo *predict* accetta in input una lista (o un array), infatti è possibile stimare più valori di y, utilizzando più x, con un'unica chiamata al metodo.

```
y_pred = model.predict([2, -2, 21, 12, -37])
print(y_pred)
```

Come vedi i valori che otteniamo non sono molto distanti a quelli che otterremo utilizzando la funzione definita sopra, questo vuol dire che siamo riusciti ad approssimarla abbastanza bene. Per vedere i valori di a (peso) e b (bias) che la funzione ha appreso possiamo usare l'attributo weights.

```
model.weights
```

La funzione che il modello ha appreso è la seguente:
$$y = 1.97x + 1.38$$
