# Quanto vale la mia App ?

```
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```
SCALE_FACTOR = 10000
```

```
x = np.array([1000, 3000, 5000, 10000, 50000, 100000])
y = np.array([4000, 10000, 16000, 31000, 151000, 301000])
```

```
x = x/SCALE_FACTOR
y = y/SCALE_FACTOR
```

```
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```
model.compile(optimizer="sgd", loss="mean_squared_error")
```

```
model.fit(x, y, epochs=500)
```

```
my_mau = 70000
x_pred = 70000/SCALE_FACTOR

y_pred = model.predict([x_pred])

my_app_worth = np.round(y_pred * SCALE_FACTOR, 2)

print("La tua app con %d MAU vale %.2f $)" % (my_mau, my_app_worth))
```

```
w = model.weights[0].numpy()[0]
b = model.weights[1].numpy()[0]

print("Peso: %.4f" % w)
print("Bias: %.4f" % (b))
```
